```
#Profile handling
import pandas as pd
data = pd.read_csv('인지심리학_ANT_게임실력 추가_1106.csv') # get data loaded.
import csv
k = 1
with open('인지심리학_개인별_Profile_오답제거.csv', 'w',encoding = 'utf-8') as f:
        writer = csv.writer(f)
        writer.writerow(['No','국적','성별','만 나이','게임 여부','하루 평균 게임 시간','게임 장르','게임 실력',
                         'Cue-S', 'Cue-N', 'Cue-D', 'Cue-C', 'Target-C', 'Target-N', 'Target-I', 
                        'S&C','S&N','S&I','N&C','N&N','N&I','D&C','D&N','D&I','C&C','C&N','C&I'])
while k<39:
    obj = data[data['유저번호'] == k]
    cue_s = []; cue_n = []; cue_d = []; cue_c = []
    target_c = []; target_n = []; target_i = []
    s_and_c = []; s_and_n = []; s_and_i = []
    n_and_c = []; n_and_n = []; n_and_i = []
    d_and_c = []; d_and_n = []; d_and_i = []
    c_and_c = []; c_and_n = []; c_and_i = []
    for i in range(0,len(obj)):
        #cue condition
        if obj.iloc[i,2] == 'S':
            cue_s.append(int(obj.iloc[i,7]))
            if obj.iloc[i,3] == 'C':
                target_c.append(int(obj.iloc[i,7]))
                s_and_c.append(int(obj.iloc[i,7]))
            elif obj.iloc[i,3] == 'N':
                target_n.append(int(obj.iloc[i,7]))
                s_and_n.append(int(obj.iloc[i,7]))
            elif obj.iloc[i,3] == 'I':
                target_i.append(int(obj.iloc[i,7]))
                s_and_i.append(int(obj.iloc[i,7]))
        elif obj.iloc[i,2] == 'N':
            cue_n.append(int(obj.iloc[i,7]))
            if obj.iloc[i,3] == 'C':
                target_c.append(int(obj.iloc[i,7]))
                n_and_c.append(int(obj.iloc[i,7]))
            elif obj.iloc[i,3] == 'N':
                target_n.append(int(obj.iloc[i,7]))
                n_and_n.append(int(obj.iloc[i,7]))
            elif obj.iloc[i,3] == 'I':
                target_i.append(int(obj.iloc[i,7]))
                n_and_i.append(int(obj.iloc[i,7]))
        elif obj.iloc[i,2] == 'D':
            cue_d.append(int(obj.iloc[i,7]))
            if obj.iloc[i,3] == 'C':
                target_c.append(int(obj.iloc[i,7]))
                d_and_c.append(int(obj.iloc[i,7]))
            elif obj.iloc[i,3] == 'N':
                target_n.append(int(obj.iloc[i,7]))
                d_and_n.append(int(obj.iloc[i,7]))
            elif obj.iloc[i,3] == 'I':
                target_i.append(int(obj.iloc[i,7]))
                d_and_i.append(int(obj.iloc[i,7]))
        elif obj.iloc[i,2] == 'C':
            cue_c.append(int(obj.iloc[i,7]))
            if obj.iloc[i,3] == 'C':
                target_c.append(int(obj.iloc[i,7]))
                c_and_c.append(int(obj.iloc[i,7]))
            elif obj.iloc[i,3] == 'N':
                target_n.append(int(obj.iloc[i,7]))
                c_and_n.append(int(obj.iloc[i,7]))
            elif obj.iloc[i,3] == 'I':
                target_i.append(int(obj.iloc[i,7]))
                c_and_i.append(int(obj.iloc[i,7]))
    avg_cue_s = sum(cue_s)/len(cue_s)
    avg_cue_n = sum(cue_n)/len(cue_n)
    avg_cue_d = sum(cue_d)/len(cue_d)
    avg_cue_c = sum(cue_c)/len(cue_c)
    avg_tar_c = sum(target_c)/len(target_c)
    avg_tar_n = sum(target_n)/len(target_n)
    avg_tar_i = sum(target_i)/len(target_i)
    avg_s_c = sum(s_and_c)/len(s_and_c)
    avg_s_n = sum(s_and_n)/len(s_and_n)
    avg_s_i = sum(s_and_i)/len(s_and_i)
    avg_n_c = sum(n_and_c)/len(n_and_c)
    avg_n_n = sum(n_and_n)/len(n_and_n)
    avg_n_i = sum(n_and_i)/len(n_and_i)
    avg_d_c = sum(d_and_c)/len(d_and_c)
    avg_d_n = sum(d_and_n)/len(d_and_n)
    avg_d_i = sum(d_and_i)/len(d_and_i)
    avg_c_c = sum(c_and_c)/len(c_and_c)
    avg_c_n = sum(c_and_n)/len(c_and_n)
    avg_c_i = sum(c_and_i)/len(c_and_i)
    with open('인지심리학_개인별_Profile_오답제거.csv', 'a', encoding = 'utf-8') as f:
        writer = csv.writer(f)
        writer.writerow([obj.iloc[0,0], obj.iloc[0,9],obj.iloc[0,10],obj.iloc[0,11],obj.iloc[0,12],obj.iloc[0,13],obj.iloc[0,14],obj.iloc[0,15],
                        avg_cue_s, avg_cue_n, avg_cue_d, avg_cue_c, avg_tar_c, avg_tar_n, avg_tar_i,
                        avg_s_c, avg_s_n, avg_s_i, avg_n_c, avg_n_n, avg_n_i, avg_d_c, avg_d_n, avg_d_i, avg_c_c, avg_c_n, avg_c_i])
    k = k+1
```
